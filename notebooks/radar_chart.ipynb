```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

np.random.seed(6)


labels = ['A', 'B', 'C', 'D', 'E']
N = len(labels)
angles = np.linspace(0, 2 * np.pi, N, endpoint=False).tolist()
angles += angles[:1]  # close the polygon

data = np.random.rand(3, N)

fig, ax = plt.subplots(figsize=(4, 4), subplot_kw=dict(polar=True))

colors = ['#FF9800', '#2196F3', '#9C27B0']
for i, row in enumerate(data):
    row = row
    values = row.tolist() + row[:1].tolist()
    ax.plot(angles, values, color=colors[i], linewidth=3)
    ax.fill(angles, values, color=colors[i], alpha=0.7)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(['' for i in labels], fontsize=11)
ax.set_yticks([])
# ax.set_yticklabels(['0.25', '0.5', '0.75', '1.0'], fontsize=7, color='grey')
ax.set_ylim(0, 1)
ax.set_theta_zero_location('N')
ax.set_theta_direction(-1)

plt.tight_layout()
out = Path('/Users/magdalenalederbauer/Code/MixSense/notebooks/radar_chart.svg')
plt.savefig(out, format='svg', bbox_inches='tight')
plt.show()
```
